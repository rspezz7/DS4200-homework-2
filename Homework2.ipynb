### Part 3 - Basic data analysis with Python

Here is a dataset of Seattle weather in 4 years. Load the data and generate corresponding data visualizations. Make sure each figure has title, labels and legends if necessary. Add one or two sentences in each figure to talk about what you have learned from the figure. 

```python
from vega_datasets import data

weather = data.seattle_weather()
weather.head()
```

##### Question 1 (5 pts)

Generate a data visualization to show the relationship between precipitation and max temperature.  

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.scatter(weather['precipitation'], weather['temp_max'], alpha = 0.7)

plt.xlabel('Precipitation')
plt.ylabel('Maximum Temperature')
plt.title('Scatter Plot of Precipitation and Maximum Temperature in Seattle')

plt.show()
```

The visualization shows that the maximum temperature in Seattle seems to stay within the range of 0 and 30 degrees, no matter the precipitation amount. The highest max temperature was recorded when there was no precipitation, and the most data is recorded when there is no precipitation.

##### Question 2 (5 pts)

Generate a data visualization to show the relationship between max temperature and weather type.  

```python
sns.boxplot(x = 'weather', y = 'temp_max', data = weather)

plt.xlabel('Weather')
plt.ylabel('Maximum Temperature')
plt.title('Maximum Temperature by Weather Type in Seattle')

plt.show()
```

The visualization shows that of all the weather types, the highest temperature is reached when it's sunny, the lowest max temperature is when it's snowing, and the biggest range of max temperatures is when it's drizzling. 

##### Question 3 (5 pts)

Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type. 

```python
sns.scatterplot(x = 'temp_max', y = 'temp_min', hue = 'weather', alpha = 0.7, data = weather)

plt.xlabel('Maximum Temperature')
plt.ylabel('Minimum Temperature')
plt.title('Maximum and Minimum Temperatures on a day in Seattle by Weather Type')

plt.show()
```

The graph shows that a lot of the lowest minimum temperatures were reached when it was sunny. In general, the higher the maximum temperature, the higher the minimum temperature, as they seem to correlate.

##### Question 4 (5 pts) 

Generate a data visualization to show the relationship between max temperature, min temperature and wind speed. 

```python
sns.scatterplot(x = 'temp_max', y = 'temp_min', hue = 'wind', alpha = 0.7, data = weather)

plt.xlabel('Maximum Temperature')
plt.ylabel('Minimum Temperature')
plt.title('Maximum and Minimum Temperatures on a day in Seattle by Wind Speed')

plt.show()
```

The lower wind speeds seem to have the highest minimum and maximum temperatures. The higher winder speeds have mid range temperatures.

Recreate the figure in Part 3 Question 2 with Altair.

##### Question 3 (10 points)

Create a facet figure with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python
alt.Chart(weather).mark_point().encode(
    alt.X('temp_max:Q', title = 'Maximum Temperature'),
    alt.Y('temp_min:Q', title = 'Minimum Temperature'),
    alt.Color('wind:Q'),
    alt.OpacityValue(0.5),
    alt.Tooltip(['temp_max:Q', 'temp_min:Q', 'wind:O']),
).properties(width = 135, height = 135
            ).facet(
                facet = 'weather:N',
                columns = 3,
                title = 'The Relationship Between Maximum and Minimum Temperatures in Seattle'
            )
```

The visualization shows that sunny days have the most recorded data and the widest range of maximum and minimum temperatures. Snow has the least data and a low range, most of the temperatures are very low. In general, all of the plots show that higher maximum temperature means higher minimum temperature, and same with low temperatures.

##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Include necessary title and labels.

```python
weather['year'] = weather['date'].dt.year

line_max = alt.Chart(weather).mark_line(
).encode(
    alt.Y('average(temp_max):Q', title = 'Average Temperature'),
    alt.X('year', title = 'Year'),
    color=alt.value('red')
)

line_min = alt.Chart(weather).mark_line(
).encode(
    alt.Y('average(temp_min):Q'),
    alt.X('year'),
)

point_max = alt.Chart(weather).mark_circle(
).encode(
    alt.Y('average(temp_max):Q'),
    alt.X('year'),
    alt.Tooltip(['average(temp_max)']),
    color=alt.value('red')
)

point_min = alt.Chart(weather).mark_circle(
).encode(
    alt.Y('average(temp_min):Q'),
    alt.X('year'),
    alt.Tooltip(['average(temp_min)'])
)

chart_max = line_max+point_max
chart_min = line_min+point_min

chart_max+chart_min
```

This visualization shows that the maximum temperatures are all within 4 degrees of each other, and the same goes for the minimum temperatures throughout the 4 years.
